```python
# 성과정보 필요한 컬럼으로 전환해서 저장

import pandas as pd

for i in range(7,13):
    df = pd.read_csv(f'./원본/2018{i:02d}_성과정보.csv', encoding='utf-8')

    columns = [
    '기준년월',
    '발급회원번호',
    '증감율_이용건수_신용_전월',
    '증감율_이용금액_신용_전월',
    '증감율_이용건수_신용_분기',
    '증감율_이용금액_신용_분기',
    '변동률_잔액_B1M'
    ]
    
    df_filtered = df[columns]

    df_filtered.to_csv(f'2018{i:02d}_성과정보_전처리.csv', index=False, encoding='utf-8')

    print(f"{i}월 저장 완료")
```

```python
import pandas as pd

df_list = []

for i in range(7,13):
    df = pd.read_csv(f'./2018{i:02d}_성과정보_전처리.csv', encoding='utf-8')
    df_list.append(df)

df_combined = pd.concat(df_list, axis=0, ignore_index=True)

df_combined.to_csv(f'2018_성과정보_통합.csv', index=False, encoding='utf-8')
```
