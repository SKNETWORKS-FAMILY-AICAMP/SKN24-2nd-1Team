### y데이터 구상

##### 1. 이용금액_신용_B0M이 0일 때 이탈이라 하자

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 1. 기준년월 리스트 추출 및 정렬
months = sorted(df['기준년월'].unique())
m_jul, m_aug, m_sep = months[0], months[1], months[2]

# 2. 피벗 테이블 생성
df_amt = df.pivot(index='발급회원번호', columns='기준년월', values='이용금액_신용_B0M')
df_elapsed = df.pivot(index='발급회원번호', columns='기준년월', values='이용후경과월_신용')

# 3. 타겟 고객 그룹 필터링
# (1) 7월 이용금액 0 & 7월 이용후경과월 0
cond_1 = (df_amt[m_jul] == 0) & (df_elapsed[m_jul] == 0)
group_1 = df_amt[cond_1]

# (2) 8월에 '처음' 0 (7월 != 0, 8월 == 0)
cond_2 = (df_amt[m_jul] != 0) & (df_amt[m_aug] == 0)
group_2 = df_amt[cond_2]

# (3) 9월에 '처음' 0 (7, 8월 != 0, 9월 == 0)
cond_3 = (df_amt[m_jul] != 0) & (df_amt[m_aug] != 0) & (df_amt[m_sep] == 0)
group_3 = df_amt[cond_3]

# 4. 시각화 함수 정의
def plot_recovery_trend(group_data, title):
    if group_data.empty:
        print(f"[{title}] 조건에 해당하는 고객이 없습니다.")
        return
        
    plt.figure(figsize=(10, 5))
    
    # 최대 100명 샘플링하여 추세화
    sample_size = min(100, len(group_data))
    sample_data = group_data.sample(n=sample_size, random_state=42)
    
    # 개별 고객의 추이 (투명한 회색 선)
    for _, row in sample_data.iterrows():
        plt.plot(months, row.values, color='grey', alpha=0.15)
        
    # 해당 그룹 전체 추세선
    plt.plot(months, group_data.mean().values, color='red', linewidth=3, label='전체 평균 추이')
    
    plt.title(f"{title} (N={len(group_data):,})")
    plt.xlabel('기준년월')
    plt.ylabel('이용금액_신용_B0M')
    
    # x축 눈금을 문자열로 변환하여 출력 (에러 방지)
    plt.xticks(months, [str(m) for m in months]) 
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# 5. 그래프 출력 실행
plot_recovery_trend(group_1, "(1) 7월 첫 0 도달 (이용후경과월=0) 추이")
plot_recovery_trend(group_2, "(2) 8월 첫 0 도달 추이")
plot_recovery_trend(group_3, "(3) 9월 첫 0 도달 추이")
```

```python
# 7월에 이용후경과월 = 0 도 포함시켜야 하는 EU >>> 보면 33,858건이 잡힘

# 3. 타겟 고객 그룹 필터링 (1번 조건 수정)
# 7월 이용금액이 0인 모든 고객 (이용후경과월 조건 제외)
cond_1_updated = (df_amt[m_jul] == 0)
group_1_updated = df_amt[cond_1_updated]

# 4. 시각화 함수 실행 (기존 함수 재사용)
plot_recovery_trend(group_1_updated, "(1) 7월 이용금액 0원 고객 추이")
```

```python
def print_recovery_stats(group_data, zero_month_idx, title):
    """
    group_data: 분석 대상 그룹의 피벗 데이터프레임
    zero_month_idx: 처음 0을 찍은 월의 인덱스 (7월=0, 8월=1, 9월=2)
    """
    total_users = len(group_data)
    
    if total_users == 0:
        print(f"[{title}] 조건에 해당하는 고객이 없습니다.\n")
        return

    # 처음 0을 찍은 '다음 달'부터 마지막 달(12월)까지의 컬럼 리스트 추출
    after_months = months[zero_month_idx + 1 :]
    
    # 이후 기간의 데이터가 모두 0인지 확인 (하나라도 0이 아니면 False 반환)
    is_continuous_zero = (group_data[after_months] == 0).all(axis=1)
    
    # 건수 계산
    dead_count = is_continuous_zero.sum()
    recovered_count = total_users - dead_count
    
    # 비중 계산
    dead_ratio = (dead_count / total_users) * 100
    recovered_ratio = (recovered_count / total_users) * 100
    
    print(f"■ {title} (총 {total_users:,}명)")
    print(f" - 연말까지 연속 0원 유지 (완전 이탈): {dead_count:,}명 ({dead_ratio:.1f}%)")
    print(f" - 이후 한 번이라도 결제 발생 (회생): {recovered_count:,}명 ({recovered_ratio:.1f}%)\n")

# 함수 실행 (months 리스트가 [201807, 201808, 201809, 201810, 201811, 201812]로 정의되어 있다고 가정)
print_recovery_stats(group_1, 0, "7월 첫 0 도달 그룹")
print_recovery_stats(group_2, 1, "8월 첫 0 도달 그룹")
print_recovery_stats(group_3, 2, "9월 첫 0 도달 그룹")
```

```python
# 1. 타겟 고객 그룹 필터링 (이용후경과월_신용 == 2 기준)
# 조건 판별은 df_elapsed로 하고, 데이터 추출은 df_amt에서 수행합니다. >>> 이거 때문에 아래 수치가 좀 이상한감..?

# (1) 7월에 이용후경과월이 2인 고객
cond_elapsed_1 = (df_elapsed[m_jul] == 2)
group_elapsed_1 = df_amt[cond_elapsed_1]

# (2) 8월에 이용후경과월이 2인 고객
cond_elapsed_2 = (df_elapsed[m_aug] == 2)
group_elapsed_2 = df_amt[cond_elapsed_2]

# (3) 9월에 이용후경과월이 2인 고객
cond_elapsed_3 = (df_elapsed[m_sep] == 2)
group_elapsed_3 = df_amt[cond_elapsed_3]

# 2. 시각화 함수 실행 (기존 함수 재사용)
plot_recovery_trend(group_elapsed_1, "(1) 7월 이용후경과월=2 고객 추이")
plot_recovery_trend(group_elapsed_2, "(2) 8월 이용후경과월=2 고객 추이")
plot_recovery_trend(group_elapsed_3, "(3) 9월 이용후경과월=2 고객 추이")
```

데이터 정합성을 검증하기 위한 코드 (이용후경과월_신용=1 이면, 이전 달은 이용금액_신용_B0M=0이고, 2달 전은 0이 아니어야 할 것 같음. 100% 이렇게 되는 데이터셋인지 검증)

```python
for t in range(2, len(months)):
    target_month = months[t]
    prev_1_month = months[t-1]  # 1달 전
    prev_2_month = months[t-2]  # 2달 전
    
    cond = (df_elapsed[target_month] == 1)
    target_df = df_amt[cond]
    
    total_cases = len(target_df)
    
    if total_cases == 0:
        print(f"■ {target_month}월: 이용후경과월이 1인 데이터 없음\n")
        continue
        
    is_prev_1_zero = (target_df[prev_1_month] == 0)
    is_prev_2_not_zero = (target_df[prev_2_month] != 0)
    
    valid_cases = (is_prev_1_zero & is_prev_2_not_zero).sum()
    error_cases = total_cases - valid_cases
    
    print(f"■ {target_month}월 데이터 검증 (대상: {total_cases:,}건)")
    print(f" - 정상 (이론과 일치): {valid_cases:,}건 ({(valid_cases/total_cases)*100:.2f}%)")
    print(f" - 오류 (이론과 불일치): {error_cases:,}건")
    print("-" * 40)

# 피곤해서 잘은 안 봤는데.. 저게 찐 오류인건지 아니면 아닐수도 있는 요소가 있긴 한 건지 고민해보긴 해야할 듯...
# 이용후경과월_신용 = 1 은 사실 1이 며칠부터 며칠까지 1달로 간주되는지를 모르니까 오차..?가 조금은 있을 수 있을 거 같다.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 1. 기준년월 리스트 추출 및 7월 변수 할당
months = sorted(df['기준년월'].unique())
m_jul = months[0] # 7월 (예: 201807)

# 2. 조건 판별용(R3M)과 시각화용(B0M) 피벗 테이블 각각 생성
df_r3m = df.pivot(index='발급회원번호', columns='기준년월', values='이용금액_R3M_신용')
df_b0m = df.pivot(index='발급회원번호', columns='기준년월', values='이용금액_신용_B0M')

# 3. 조건 필터링: 7월의 R3M 값이 0인 발급회원번호 추출
cond_r3m_zero = (df_r3m[m_jul] == 0)

# 4. 시각화할 타겟 데이터 추출: 해당 회원들의 B0M 전체(7월~12월) 데이터
target_b0m_trend = df_b0m[cond_r3m_zero]

# 5. 시각화
if target_b0m_trend.empty:
    print("7월 이용금액_R3M_신용이 0인 고객이 없습니다.")
else:
    plt.figure(figsize=(10, 5))
    
    # 렌더링 최적화를 위해 최대 100명 무작위 샘플링
    sample_size = min(100, len(target_b0m_trend))
    sample_data = target_b0m_trend.sample(n=sample_size, random_state=42)
    
    # 샘플링된 개별 고객의 B0M 추이 (투명한 회색 선)
    for _, row in sample_data.iterrows():
        plt.plot(months, row.values, color='grey', alpha=0.15)
        
    # 해당 그룹 전체의 평균 B0M 추세선 (굵은 빨간색 선)
    plt.plot(months, target_b0m_trend.mean().values, color='red', linewidth=3, label='그룹 평균 B0M')
    
    plt.title(f"7월 R3M=0 고객의 당월 이용금액(B0M) 추이 (N={len(target_b0m_trend):,})")
    plt.xlabel('기준년월')
    plt.ylabel('이용금액_신용_B0M')
    
    # x축 눈금을 문자열로 변환하여 출력
    plt.xticks(months, [str(m) for m in months]) 
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
# target_b0m_trend: 앞선 코드에서 추출한 '7월 R3M=0' 고객들의 B0M 피벗 데이터
# months: ['201807', '201808', '201809', '201810', '201811', '201812'] (예시)

total_users_r3m = len(target_b0m_trend)

if total_users_r3m == 0:
    print("7월 R3M=0 조건에 해당하는 고객이 없습니다.\n")
else:
    # 7월의 다음 달인 8월(인덱스 1)부터 12월까지의 컬럼 추출
    after_months = months[1:]
    
    # 8~12월 기간 동안 B0M 이용금액이 모두 0인지 확인 (하나라도 0이 아니면 False)
    is_continuous_zero = (target_b0m_trend[after_months] == 0).all(axis=1)
    
    # 완전 이탈(연속 0원) 및 회생(1회 이상 결제) 건수 계산
    dead_count = is_continuous_zero.sum()
    recovered_count = total_users_r3m - dead_count
    
    # 비중 계산
    dead_ratio = (dead_count / total_users_r3m) * 100
    recovered_ratio = (recovered_count / total_users_r3m) * 100
    
    print(f"■ 7월 R3M=0 고객 그룹 B0M 추적 결과 (총 {total_users_r3m:,}명)")
    print(f" - 연말까지 당월 결제액 연속 0원 유지 (완전 이탈): {dead_count:,}명 ({dead_ratio:.1f}%)")
    print(f" - 8~12월 중 당월 결제액 1회 이상 발생 (회생): {recovered_count:,}명 ({recovered_ratio:.1f}%)\n")
```
