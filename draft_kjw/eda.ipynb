### 이탈 정의
- vip_df['churn'] = (vip_df['next_usage'] == 0).astype(int)
- 이번 달 기준 -> 다음 달에 사용 안 할 확률
    - next_usage=0 -> 거의 이탈

```python
# 이용금액 감소율 분포
# 'next_usage=다음 달 이용금액' 만듦
df = df.sort_values(['발급회원번호','기준년월'])        # 같은 고객을 시간 순서대로 정렬: sort
df['next_usage'] = df.groupby('발급회원번호')['이용금액_신용_B0M'].shift(-1)        
(df['next_usage']==0).mean()


# [참고] groupby: 고객별로 묶음 
# [참고] shift(-1): 한 칸 위 값 가져오기(미래) <-> shift(1): 한 칸 아래 값 가져오기 (과거) 
```

---

```python
# VIP 고객만 필터
vip_df = df[df['VIP등급코드'].notna()]
vip_df
```

```python
# [참고] 등급 코드
#01~03 : VVIP/ 04~10 : VIP

#04 : VIP_4등급
#05 : VIP_3등급
#06 : VIP_2등급
#07 : VIP_1등급

#전체 고객 대상 churn 예측 모델 구축
```

```python
# csv 읽어온 df 복사본 생성 (모델 생성용)
df = df.copy()
df['churn'] = (df['이용후경과월_신용'] >= 2).astype(int)
```

```python
# [참고] 제거 이유
# 1) 발급회원번호: 식별자 (패턴 학습에 의미 없음)
# 2) 기준년월: 시간 정보 그대로 넣으면 leakage 위험
# 3) 이용후경과월_신용: 타겟 생성 기준
# 4) 탈회횟수_누적: 이미 탈회 결과값
# 5) 최종탈회후경과월: 결과 기반
# 6) 탈회횟수_발급1년이내: 결과 기반
drop_cols = [
    'Unnamed: 0',
    '발급회원번호',
    '기준년월',
    '이용후경과월_신용',  # 타겟 기준
    '탈회횟수_누적',
    '최종탈회후경과월',
    '탈회횟수_발급1년이내',
    'next_usage'
]

df_model = df.drop(columns=drop_cols)
```

```python
# X, y 데이터 분리
X = df_model.drop(columns=['churn'])
y = df_model['churn']
```

```python
X.dtypes
```

```python
# [참고] 원핫 인코딩
# 연령, 거주시도명 등 범주형 타입 처리 (문자열 계산 불가이기 때문)
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# stratify (이탈 비율 유지하기 위함)
# 타겟 비율 유지한 채 학습/검증 데이터 분리
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=300,
    max_depth=10,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
# 예측
# [참고] X = churn을 제외한 모든 변수
# [참고] 'X, y 데이터' 분리 참고하면 됨
y_pred = model.predict(X_test)
```

```python
# 평가
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
# [참고] 성능이 너무 좋게 나와서 next_usage를 drop하고 다시 수정
```

### 📌 *[참고]* '평가' 해석
##### class 0 = 정상 고객
- precision 0.97: 정상이라고 예측한 것 중 97%가 진짜 정상
- recall 0.95: 실제 정상 고객의 95%를 맞춤
- f1 0.96 : 거의 완벽

##### class 1 = 이탈 고객 
- precision 0.63: 이탈이라고 예측한 고객 중 63%만 진짜 이탈
- recall 0.72: 실제 이탈 고객의 72%를 잡아냄
- f1 0.68: 이탈 탐지 성능 종합 점수
