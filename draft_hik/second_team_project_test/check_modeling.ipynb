````python
feature_cols = [
    # 고객 속성
    "연령",
    "Life_Stage",
    "VIP등급코드",
    "거주시도명",

    # 상태 정보
    "회원여부_이용가능",
    "회원여부_연체",
    "탈회횟수_누적",
    "탈회횟수_발급1년이내",
    "최종탈회후경과월",

    # 리스크 / 한도
    "강제한도감액금액_R12M",
    "한도증액금액_R12M",
    "연체일수_B2M",

    # 잔액 / 평잔
    "잔액_B0M",
    "월중평잔",
    "평잔_3M",
    "평잔_6M",

    # 변화율
    "변동률_잔액_B1M"
]

X = df[feature_cols]
y = df["y"]
````

````python
# 1. 시간 기준 분리
train = df[df["기준년월"] <= 201811]
test  = df[df["기준년월"] == 201812]

# 2. X / y 분리
X_train = train[feature_cols]
y_train = train["y"]

X_test  = test[feature_cols]
y_test  = test["y"]
````

````python
X_train["Life_Stage"] = X_train["Life_Stage"].astype("category").cat.codes
X_test["Life_Stage"]  = X_test["Life_Stage"].astype("category").cat.codes
X_train["거주시도명"] = X_train["거주시도명"].astype("category").cat.codes
X_test["거주시도명"]  = X_test["거주시도명"].astype("category").cat.codes
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

dt = DecisionTreeClassifier(max_depth=7, random_state=42)
dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
````

## 랜덤포레스트

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from imblearn.over_sampling import SMOTE
X = df[feature_cols]
y = df["y"]

# ===========================================
# 2️⃣ 시간 기준으로 train/test 분리
# ===========================================
train = df[df["기준년월"] <= 201811]
test  = df[df["기준년월"] == 201812]

X_train = train[feature_cols]
y_train = train["y"]

X_test  = test[feature_cols]
y_test  = test["y"]

# ===========================================
# 3️⃣ SMOTE로 학습용 데이터 증강
# ===========================================
smote = SMOTE(random_state=42)
X_train_smt, y_train_smt = smote.fit_resample(X_train, y_train)
````

````python
# ===========================================
# 4️⃣ 랜덤포레스트 학습
# ===========================================
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced',  # 불균형 클래스 고려
    max_depth=10
)
rf.fit(X_train_smt, y_train_smt)

# ===========================================
# 5️⃣ 평가
# ===========================================
y_pred = rf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
````

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from imblearn.over_sampling import SMOTE

# 1️⃣ ID 컬럼 제거
# df_09 = df_09.drop(columns=['발급회원번호'])

# 1️⃣ 먼저 문자열로 변환
df_09['이용금액대'] = df_09['이용금액대'].astype(str)

# 2️⃣ 
# 숫자가 아닌 값은 NaN으로 바꾸고
# NaN을 0으로 처리
df_09['이용금액대'] = pd.to_numeric(
    df_09['이용금액대'].str.replace('만원','')
                          .str.replace('+',''), errors='coerce'
)

# 3️⃣ NaN은 0으로 채우기
df_09['이용금액대'] = df_09['이용금액대'].fillna(0)

# 3️⃣ X, y 정의
X = df_09.drop(columns=['y', '발급회원번호', '이용건수_신용_B0M', '이용금액_신용_B0M', '이용후경과월_신용', '이용금액_신용_R12M'], axis=1)
y = df_09['y']

# 4️⃣ 범주형 컬럼 선택
categorical_cols = ['거주시도명', 'Life_Stage']

# 5️⃣ 학습/테스트 분리
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0, stratify=y
)

# 6️⃣ One-Hot Encoder 학습
ohe = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
ohe.fit(X_train[categorical_cols])

# 7️⃣ 범주형 컬럼 변환
X_train_encoded = pd.DataFrame(ohe.transform(X_train[categorical_cols]),
                               columns=ohe.get_feature_names_out(categorical_cols),
                               index=X_train.index)
X_test_encoded = pd.DataFrame(ohe.transform(X_test[categorical_cols]),
                              columns=ohe.get_feature_names_out(categorical_cols),
                              index=X_test.index)

# 8️⃣ 원래 숫자 컬럼과 합치기
num_cols = X_train.drop(columns=categorical_cols).columns
X_train_final = pd.concat([X_train[num_cols], X_train_encoded], axis=1)
X_test_final = pd.concat([X_test[num_cols], X_test_encoded], axis=1)

smote = SMOTE(random_state=0)
X_train_res, y_train_res = smote.fit_resample(X_train_final, y_train)

# 9️⃣ RandomForest 학습
rf = RandomForestClassifier(max_depth=5, random_state=0)
rf.fit(X_train_res, y_train_res)

# 10️⃣ 평가
y_pred = rf.predict(X_test_final)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
````
