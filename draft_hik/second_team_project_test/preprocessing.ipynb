### 변동되었는지를 판단
- 10월에 1이 변동이었으면 11 ~ 12에도 1 유지

```python
# 월 기준 정렬
data2 = data.sort_values(["발급회원번호", "기준년월"])

# 전월 대비 탈회횟수 증가 여부
chg = data2.groupby("발급회원번호")["탈회횟수_누적"].diff() == 1

# 탈회 변동이 있었던 회원번호 (중복 제거)
churn_members = data2.loc[chg, "발급회원번호"].unique()
```

```python
data.loc[
    data["발급회원번호"].isin(churn_members),
    "탈회횟수_누적"
] += 1
```

```python
data[data["발급회원번호"].isin(churn_members)].groupby("발급회원번호")["탈회횟수_누적"].unique()
```

# 이용금액대 전처리

```python
result = (
    data[["발급회원번호", "연령", "Life_Stage"]]
    .drop_duplicates()
)

result.to_csv("LifeStage측정.csv", index=False, encoding="UTF-8")
```

```python
base = data[["발급회원번호", "연령", "Life_Stage"]].drop_duplicates()
base

cnt = base.groupby(["연령", "Life_Stage"]).size().reset_index(name="건수")
cnt

total = cnt.groupby(["연령"])["건수"].sum().reset_index(name="전체건수")
total

result = cnt.merge(total, on="연령", how="left")
result
```

```python
base = data[["기준년월", "발급회원번호", "연령", "Life_Stage"]].drop_duplicates()
base

cnt = base.groupby(["기준년월", "연령", "Life_Stage"]).size().reset_index(name="건수")
cnt

total = cnt.groupby(["연령"])["건수"].sum().reset_index(name="전체건수")
total

result = cnt.merge(total, on="연령", how="left")
result.to_csv("LifeStage상세히.csv", index=False, encoding="UTF-8")
```
